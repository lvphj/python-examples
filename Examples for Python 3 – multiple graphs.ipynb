# Plot multiple graphs

## Plotting a histogram and a time plot side by side

### Import data

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import phjGenerateExampleData
```

```python
phjTempDF = phjGenerateExampleData.phjPandasDF9()
```

```python
print("Head")
print(phjTempDF.head(3))
print("\nTail")
print(phjTempDF.tail(3))
print("\n")
print(phjTempDF.dtypes)
```

### Plot two basic graphs

```python
plt.figure(1)    # Optional. Will be created automatically if don't specify axes.

# Graph 1
plt.subplot(121) # Optional. Will be created automatically if don't specicy axes.
n, bins, patches = plt.hist(phjTempDF['random_normal'])

# Graph 2
plt.subplot(122) # First (and in this case, only) subplot in the first figure. Optional. Will be created automatically if don't specify axes.
plt.plot_date(phjTempDF['date'], phjTempDF['random_normal'], '-')

plt.show()
```

### Adding bling

```python
plt.figure(1)    # Optional. Will be created automatically if don't specify axes.

# Graph 1 - histogram
plt.subplot(121) # Optional. Will be created automatically if don't specicy axes.
n, bins, patches = plt.hist(phjTempDF['random_normal'])
plt.title('Histogram')
plt.ylabel('Frequency')
plt.xlabel('Random variable')
plt.grid(True)

# Graph 2 - time-series
plt.subplot(122) # First (and in this case, only) subplot in the first figure. Optional. Will be created automatically if don't specify axes.
plt.plot_date(phjTempDF['date'], phjTempDF['random_normal'], '-')
plt.title('Time series')
plt.ylabel('Random variable')
plt.xlabel('Date')
plt.grid(True)

plt.show()
```
